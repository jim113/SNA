### Α) Δημιουργία και οπτικοποίηση σύνθετων τύπων δικτύου

Κατασκευή και οπτικοποίηση της τοπολογίας REG για βαθμό πλέγματος $d = 4$

```python
d = 4

# για την κατασκευή του REG χρησιμοποιήστε την watts_strogatz_graph από το networkx. Επιλέξτε προσεκτικά τις παραμέτρους. 
REG =  nx.watts_strogatz_graph(n, d, 0) # p = 0 because we have regular lattice

# για την οπτικοποίηση, χρησιμοποιήστε την draw_circular από το networkx
nx.draw_circular(REG)
```

Κατασκευή και οπτικοποίηση της τοπολογίας RG(ER) για αριθμό συνδέσεων $M = 750$

```python
M = 750

# για την κατασκευή του RGER χρησιμοποιήστε την gnm_random_graph από το networkx
RGER = nx.gnm_random_graph(n, M)

# για την οπτικοποίηση, χρησιμοποιήστε την draw_circular από το networkx
nx.draw_circular(RGER)
```

Κατασκευή και οπτικοποίηση της τοπολογίας RGG σε περιοχή $L\times L = 1$ (unit cube) και ακτίνα σύνδεσης $R=0.25$

```python
R = 0.25
L = 1
n = 170

# για την κατασκευή του RGG, χρησιμοποιήστε την random_geometric_graph από το networkx
RGG = nx.random_geometric_graph(n, R, dim = 2) 

# για την οπτικοποίηση, χρησιμοποιήστε draw_network_edges, draw_network_nodes.
# Για τη θέση των κόμβων στην τετράγωνη περιοχή, χρησιμοποιήστε την get_node_attributes από το networkx.
pos = nx.get_node_attributes(RGG, 'pos') 

#print(list(RGG.nodes(data=True)))
#print(pos)

nx.draw_networkx_nodes(RGG, pos, node_size=100, node_color='r')

nx.draw_networkx_edges(RGG, pos, width=0.5, edge_color='k')
```

Κατασκευή και οπτικοποίηση της τοπολογίας SF για βαθμό αρχικού πλέγματος $d=4$

```python
d = 4

# για την κατασκευή του SF, χρησιμοποιήστε την barabasi_albert_graph από το networkx.
SF = nx.barabasi_albert_graph(n, d)

# για την οπτικοποίηση, χρησιμοποιήστε την draw_circular από το networkx.
nx.draw_circular(SF)
```

Κατασκευή και οπτικοποίηση της τοπολογίας SW για βαθμό αρχικού πλέγματος $d=4$ και πιθανότητας ανασύνδεσης $p=0.3$

```python
d = 4
p = 0.3

# για την κατασκευή του SW, χρησιμοποιήστε την watts_strogatz_graph από το networkx.
SW = nx.watts_strogatz_graph(n, d, p)

# για την οπτικοποίηση, χρησιμοποιήστε την draw_circular από το networkx.
nx.draw_circular(SW)
```

### Γ) Δίκτυα με βάρη

Για ένα δίκτυο με βάρη στις συνδέσεις του, όμοια με τον πίνακα γειτνίασης, ορίζεται ο πίνακας βαρών, $W=[w_{ij}]$, όπου με $w_{ij}>0$ συμβολίζεται το βάρος στην ακμή που ενώνει τους κόμβους $i,j$. Για μη-κατευθυνόμενους γράφους ο πίνακας $W$ είναι συμμετρικός.

Για κάθε μια από τις τοπολογίες REG, RG(ER), RGG, SF(BA) και SW(WS) του πίνακα, να δημιουργηθεί τυχαίος πίνακας βαρών. Το εύρος των βαρών να κυμαίνεται στο διάστημα $[1,10]$.

```python
# για τιμές στο [1,10]: α=1, b=10, weight=np.random.rand()*(b-a) + a
# για την προσθήκη βαρών στις ακμές, χρησιμοποήστε την add.edge του networkx
#weights = RGG.edges.data('weight') 
#print(weights)
def weight_calculator(n):
    weight_line = lambda n : [np.random.rand()*(9) + 1 for i in range(n)]
    weights = np.array([weight_line(n) for i in range(n)]) # create random array of weights
    for i in range(n):
        weights[i][i] = 0
    return np.maximum(weights, weights.transpose()) # inorder to make it symmetrical

def edge_updater(G):
    weights = weight_calculator(n)
    for e in G.edges():
        G[e[0]][e[1]]['weight'] = weights[e[0]][e[1]] # update edge weight according to weights if the edge exists

#print(len(weight[0]), len(weight))
#weights = weight_calculator(n)
#print(weights)

edge_updater(REG)
edge_updater(RGER)
edge_updater(RGG)
edge_updater(SF)
edge_updater(SW)
#print(RGG.edges.data('weight'))
#for e in RGG.edges():
 #   print(RGG[e[0]][e[1]])
```

##### i) Να υπολογιστεί η δύναμη του κάθε κόμβου

```python
#δίνεται η συνάρτηση strength που έχει ως όρισμα την τοπολογία που παράξατε στο ερώτημα Α:
def strength(topology):
    weights_list=[]
    for i in range(0,len(topology.adj)):
        a=[]
        for j in range (0,len(topology.adj[i].items())):
            d=list(topology.adj[i].values())[j]['weight']            
            #c=d.values()
            a.append(d)
        weights_list.append(a)    
    node_weights = []
    for node in weights_list:
        node_weights.append(sum(node))
    return node_weights

REG_strength = strength(REG)
RGER_strength = strength(RGER)
RGG_strength = strength(RGG)
SF_strength = strength(SF)
SW_strength = strength(SW)

print('REG strength: ', REG_strength)
print('RGER strength: ', RGER_strength)
print('RGG strength: ', RGG_strength)
print('SF strength: ', SF_strength)
print('SW strength: ', SW_strength)
```

```python
# χρησιμοποήστε την all_pairs_shortest_path_length του networkx 
def compute_path_lengths(G):
    return nx.all_pairs_shortest_path_length(G)

#print([i for i in compute_path_lengths(REG)])
REG_path_lengths = compute_path_lengths(REG)
RGER_path_lengths = compute_path_lengths(RGER)
RGG_path_lengths = compute_path_lengths(RGG)
SF_path_lengths = compute_path_lengths(SF)
SW_path_lengths = compute_path_lengths(SW)

def path_lengths_listing(given_lengths): #arragne the dictionary into list in order to find mean and std
    result = []
    lengths_of_all_paths = [i[1] for i in given_lengths]
    lengths_without_destination_node = ([i.values() for i in lengths_of_all_paths])
    for i in range(len(lengths_without_destination_node)):
        for j in lengths_without_destination_node[i]:
            result.append(j) # aggregate all path lengths and append them to a list
    return result

#print(list(REG_path_lengths))
#print(np.var(path_lengths_listing(REG_path_lengths)))
REG_list_path = path_lengths_listing(REG_path_lengths)
RGER_list_path = path_lengths_listing(RGER_path_lengths)
RGG_list_path = path_lengths_listing(RGG_path_lengths)
SF_list_path = path_lengths_listing(SF_path_lengths)
SW_list_path = path_lengths_listing(SW_path_lengths)

print('Mean path length REG: ', np.mean(REG_list_path))
print('Variance of path length REG: ', np.var(REG_list_path), '\n')

print('Mean path length RGER: ', np.mean(RGER_list_path))
print('Variance of path length RGER: ', np.var(RGER_list_path), '\n')

print('Mean path length RGG: ', np.mean(RGG_list_path))
print('Variance of path length RGG: ', np.var(RGG_list_path), '\n')

print('Mean path length SF: ', np.mean(SF_list_path))
print('Variance of path length SF: ', np.var(SF_list_path), '\n')

print('Mean path length SW: ', np.mean(SW_list_path))
print('Variance of path length SW: ', np.var(SW_list_path))
```

O γενικός τύπος που θα εφαρμοστεί και μας δίνει τον τοπικό ΣΟ είναι ο εξής: <br>

$\Large C_i = 2 \cdot \frac{|e_{jk} \hspace{1mm}: \hspace{1mm} u_j,u_k \in N_i, e_{j,k} \in E|}{k_i(k_i-1)}$ <br>

<p>Topology 1<p>
    Για κάθε κόμβο του αστεριού παρατηρούμε ότι κανένας από τους γείτονες του δεν συνδέεται με κάποιον άλλο γείτονα. Συνεπώς ο αριθμητής του κλάσματος είναι μηδενικός και άρα ο τοπικός ΣΟ είναι μηδέν για κάθε κόμβο και άρα και η μέση τιμή είναι μηδέν.
<p>Topology 2<p>
    Στην τοπολογία αυτή διακρίνουμε 2 είδη κόμβων, τον κεντρικό και όλους τους περιφεριακούς αυτού, των οποίων τα χαρακτηριστικά είναι τα ίδια. Συνεπώς για κάθε εναν περιφεριακό κόμβο θα ισχύει ότι: <br>
$C_{περ} = \frac{2 \cdot 1}{2 \cdot 1}$ = 1 <br>
    Για τον κεντρικό θα ισχύει ότι έχει 4 γείτονες και αυτοί μεταξύ τους έχουν 2 ακμές άρα: <br>
    $C_{κεντρ} = \frac{2 \cdot 2}{4 \cdot 3}$ = 1/3 <br>
<p>Topology 3</p>
Εφαρμόζοντας τους ίδιους τύπους με την τοπολογία 2 θα έχουμε: <br>
Για κάθε εναν περιφεριακό κόμβο θα ισχύει ότι: <br>
$C_{περ} = \frac{2 \cdot 2}{3 \cdot 2}$ = 2/3 <br>
    Για τον κεντρικό θα ισχύει ότι έχει 4 γείτονες και αυτοί μεταξύ τους έχουν 4 ακμές άρα: <br>
    $C_{κεντρ} = \frac{2 \cdot 4}{4 \cdot 3}$ = 2/3 <br>
    
Η μέση τιμή για κάθε τοπολογία δίνει τον μέσο τοπικό ΣΟ.

<h4>Αναλυτικός υπολογισμός κεντρικοτήτων</h4>
Θα χρησιμοποιήσουμε τους ακόλουθους τύπους: <br>
Deg Centr = $\frac{k_i}{M}$ <br>
Clos Centr = Inverse of AVG of shortest paths from node to all others = 1 / AVG<br>
Between Centr = ποσοστό παρουσίας κόμβου στα ελάχιστα μονοπάτια μεταξύ όλων των υπόλοιπων κόμβων. <br>

### <p>Τοπολογία Α </p>

$Deg_1 = 3/3 = 1$, $Deg_2=Deg_3=Deg_4=1/3$ <br>

$Clos_1 = 3/(1+1+1)=1, Clos_2=3/(1+2+2)=0.6, Clos_3=3/(1+2+2)=0.6, Clos_4=3/(1+2+2)=0.6$ <br>

$Bet_1 = 3/3=1$, γιατί υπάρχει σε όλα τα shortest paths o κόμβος 1. <br>
$Bet_2=Bet_3=Bet_4=0$, γιατί στα υπόλοιπα shortest paths μεταξύ των υπόλοιπων κόμβων δεν συμμετέχει κανένας από αυτούς τους κόμβους.<br>

### <p>Τοπολογία B </p>

$Deg_1 = Deg_2=Deg_3=Deg_4=3/3=1$ <br>

$Clos_1 = Clos_2 = Clos_3 = Clos_4 = 3/(1+1+1)=1$ <br>

$Bet_1 = Bet_2=Bet_3=Bet_4=0$, γιατί το γράφημα είναι κλίκα 4 κόμβων και άρα όλα τα shortest paths αποτελούνται από την ακμή μεταξύ της αρχής και του τέλους τους. Άρα δεν επηρεάζονται από τον εξεταζόμενο κάθε φορά κόμβο.
    

##### 2) Υπολογισμός κεντρικότητας σε μεγαλύτερες συνθετικές τοπολογίες

- degree centrality
- closeness centrality
- betweenness centrality
- eigenvector centrality

```python
def calculate_degree_centrality(G):
    degree_centrality = nx.degree_centrality(G)
    return degree_centrality

def calculate_closeness_centrality(G):
    closeness_centrality = nx.closeness_centrality(G)
    return closeness_centrality

def calculate_betweenness_centrality(G):
    betweenness_centrality = nx.betweenness_centrality(G)
    return betweenness_centrality

def calculate_eigenvector_centrality(G):
    eigenvector_centrality = nx.eigenvector_centrality(G)
    return eigenvector_centrality

REG_degree_centrality = calculate_degree_centrality(REG)
REG_closeness_centrality = calculate_closeness_centrality(REG)
REG_betweenness_centrality = calculate_betweenness_centrality(REG)
REG_eigenvector_centrality = calculate_eigenvector_centrality(REG)

RGER_degree_centrality = calculate_degree_centrality(RGER)
RGER_closeness_centrality = calculate_closeness_centrality(RGER)
RGER_betweenness_centrality = calculate_betweenness_centrality(RGER)
RGER_eigenvector_centrality = calculate_eigenvector_centrality(RGER)

RGG_degree_centrality = calculate_degree_centrality(RGG)
RGG_closeness_centrality = calculate_closeness_centrality(RGG)
RGG_betweenness_centrality = calculate_betweenness_centrality(RGG)
RGG_eigenvector_centrality = calculate_eigenvector_centrality(RGG)

SF_degree_centrality = calculate_degree_centrality(SF)
SF_closeness_centrality = calculate_closeness_centrality(SF)
SF_betweenness_centrality = calculate_betweenness_centrality(SF)
SF_eigenvector_centrality = calculate_eigenvector_centrality(SF)

SW_degree_centrality = calculate_degree_centrality(SW)
SW_closeness_centrality = calculate_closeness_centrality(SW)
SW_betweenness_centrality = calculate_betweenness_centrality(SW)
SW_eigenvector_centrality = calculate_eigenvector_centrality(SW)

# για πιο εύκολο σχολιασμό, να παράξετε 1 figure για όλες τις τοπολογίες ανά κεντρικότητα
def plot_centrality(centrality, indexer, name):
    labeler = ['REG', 'RGER', 'RGG', 'SF', 'SW']
    k = 0
    plt.figure(indexer)
    plt.title(name)
    for i in centrality:
        plt.plot(np.sort(list(i.values())),np.array(range(n))/float(n), label = labeler[k])
        k = k + 1
    plt.legend()

degree_central = [REG_degree_centrality, RGER_degree_centrality, RGG_degree_centrality, SF_degree_centrality, SW_degree_centrality]
closeness_central = [REG_closeness_centrality, RGER_closeness_centrality, RGG_closeness_centrality, SF_closeness_centrality, SW_closeness_centrality]
betweenness_central = [REG_betweenness_centrality, RGER_betweenness_centrality, RGG_betweenness_centrality, SF_betweenness_centrality, SW_betweenness_centrality]
eigenvec_central = [REG_eigenvector_centrality, RGER_eigenvector_centrality, RGG_eigenvector_centrality, SF_eigenvector_centrality, SW_eigenvector_centrality]

plot_centrality(degree_central, 1, 'Degree Centrality')
plot_centrality(closeness_central, 2, 'Closeness Centrality')
plot_centrality(betweenness_central, 3, 'Betweenness Centrality')
plot_centrality(eigenvec_central, 4, 'Eigenvector Centrality')
```

Αντίθετα σε γραφήματα πλέγματα η κεντρικότητα που αφορά τα συντομότερα μονοπάτια θα είναι μικρή καθώς υπάρχουν απομακρυσμένοι κόμβοι. <br>

Στον αντίποδα σχετικά με την κεντρικότητα σε ότι αφορά την παρουσία του κόμβου σε ελάχιστα μονοπάτια υπερισχύουν τα πλέγματα καθώς οι περισσότεροι κόμβοι ενός συντομότερο μονοπατιού παίζουν κρίσιμο ρόλο στην διατήρηση του μονοπατιού αυτού ως συντομότρο, γιατί για να τους αποφύγουμε θα πρέπει να διασχίσουμε περισσότερες ακμές και άρα το μονοπάτι να πάψει να είναι το συντομότερο δυνατό. <br>

Τέλος σε ότι αφροά την κεντρικότητα με βάση τους γείτονες κάθε κόμβου παρατηρούμε ότι τα γραφήματα με hubs οδηγούν σε μεγαλύτερη κεντρικότητα καθώς υπάρχουν πολλοί κόμβοι μικρού βαθμού συνδεδεμένοι στα hubs τα οποία θεωρούνται κεντρικοί κόμβοι και άρα αυξάνουν την κεντρικότητα των γειτόνων τους. Αντίθετα για γραφήματα τύπου πλέγαμτος η κεντρικότητα παραμένει σταθερή και πάλι καθώς οι γείτονες των κόμβων ομοιάζουν μεταξύ τους και δεν επηρεάζονται από κάποιο ιδιαίτερο χαρακτηριστικό του γραφήματος, πέραν της σταθερής δομής του.

Για τον ίδιο λόγο εμφανίζει μικρή κεντρικότητα σε ότι αφορά την παρουσία κόμβου σε συντομότερο μονοπάτι καθώς λόγω του giant component μπορούν να υπάρξουν περισσότερα του ενός συντομότερα μονοπάτια.<br>

Για την τοπολογία RGG: Για τους ίδιους λόγους με παραπάνω οδηγούμαστε σε παρόμοια συμπεράσματα τα οποία ενισχύονται από το γεγονός της γεωμετρικότητας και της θέσης των κόμβων που επηρεάζει την μεταξύ τους διασύνδεση.<br>

Για την τοπολογία SF: Αξιοποιώντας την scale free ιδιότητα καταλαβαίνουμε ότι η κεντρικότητα που αφορά τη μέση τιμή των συντομότερων μονοπατιών θα είναι η πιο αυξημένη. Αυτό συμπεραίνεται γιατί τα hubs βοηθούν στην μείωση του μήκους του συντομότερου μονοπατιού. Αντίστοιχα η μικρότερη τιμή κεντρικότητας αναμένετε να συναντάται στην ενδιαμεσική κεντρικότητα καθώς μόνο ελάχιστη κόμβοι, τα hubs, εμφανίζονται στα περισσοτερα ελαχιστα μονοπάτια ενώ στον αντίποδα η πλειονότητα των κόμβων δεν βρίσκεται στα ελάχιστα μονοπάτια μεταξύ των υπόλοιπων κόμβων.<br>

Για κάθε έναν τύπο δικτύου, να παράξετε 100 διαφορετικές τοπολογίες και να παρουσιάσετε το διάγραμμα ποσοστού συνεκτικότητας σε σχέση με τις καθορισμένες παραμέτρους κάθε δικτύου.

```python
#για τον έλεγχο της συνεκτικότητας, χρησιμοποίηστε την is_connected του networkx
#για τη γραφική αναπαράσταση των αποτελεσμάτων, μπορείτε να χρησιμοποίησετε την bar του matplotlib
#έλεγχος συνεκτικότητας για REG
def REG_conncted_percent(d, n):
    connected = 0
    for i in range(100): # create 100 graphs
        tmp_REG = nx.watts_strogatz_graph(n, d, 0)
        connected = connected + nx.is_connected(tmp_REG)
    return connected / 100.0
        
    
#plot example:
def plot_connect_REG(n, indexer):
    plt.figure(indexer)
    y = [REG_conncted_percent(d, n) for d in range(2,12,2)]
    plt.bar(list(range(2,12,2)),y, width=0.5, color='b')
    plt.title("REG n = %d" %n)
    plt.ylabel("Number of connected graphs")
    plt.xlabel("Degree")
    plt.show()
    
plot_connect_REG(100, 1)
plot_connect_REG(200, 2)
```

Σχολιάστε τα αποτελέσματα συνεκτικότητας για το REG:

Παρατηρούμε το αναμενόμενο, δηλαδή η συνεκτικότητα παραμένει στο 100% ανεξαρτήτως του πλήθους των κόμβων και του βαθμού αυτών. Το γεγονός αυτό είναι άμεση συνέπεια της τοπολογίας του δικτύου το οποίο είναι ένα κλασικό πλέγμα, κατεξοχίν συνδεδεμένο γράφημα.

```python
#έλεγχος συνεκτικότητας για RGER
def RGER_conncted_percent(M, n):
    connected = 0
    for i in range(100): # create 100 graphs
        tmp_RGER = nx.gnm_random_graph(n, M)
        connected = connected + nx.is_connected(tmp_RGER)
    return connected / 100.0
        
    
#plot example:
def plot_conncet_RGER(n, indexer):
    plt.figure(indexer)
    y = [RGER_conncted_percent(M, n) for M in range(100,900,100)]
    print(y)
    plt.bar(list(range(100,900,100)),y, width=50, color='b')
    plt.title("RGER n = %d" %n)
    plt.ylabel("Number of connected graphs")
    plt.xlabel("Edges")
    plt.show()
    
plot_conncet_RGER(100, 1)
plot_conncet_RGER(200, 2)
```

Σχολιάστε τα αποτελέσματα συνεκτικότητας για το RGER:

Για το τυχαίο γράφημα παρατηρούμε ότι εξαρτάται σε μεγάλο βαθμό η συνεκτικότητα του από το πλήθος των κόμβων καθώς αυτό επηρεάζει και το πλήθος των δυνατών τυχαίων γραφημάτων από τα αποία μπορεί να διαλέξει ο αλγόριθμος. Συγκεκριμένα παρατηρούμε ότι όσο περισσότερες είναι οι ακμές τόσο μεγαλύτερο είναι το ποσοστό των συνδεδεμένων γραφημάτων. Συγκεκριμένα για 100 κορυφές απαιτούνται τουλάχιστον 500 ακμές, από αυτή τη πληθικότητα και πάνω παρατηρούμε 100% συνδεδεμένα γραφήματα. Αντίστοιχα ανεβάζοντας το πλήθος των κορυφών παρατηρούμε άνοδο στο πλήθος των απαιτούμενων ακμών ώστε να προσεγγίσουμε πλήρος συνδεδεμένα γραφήματα. Θεωρητικά αν αυξάναμε και άλλο τις ακμές θα πετυχαίναμε και σε αυτό το γράφημα 100% συνδεδεμένα γραφήματα. Αυτό οφείλεται στην τυχαιότητα που επιλέγεται το γράφημα και στη σχέση μεταξύ πλήθους κορυφών και ακμών.

```python
#έλεγχος συνεκτικότητας RG Gilbert
#για την κατασκευή της RG χρησιμοποιήστε τη συνάρτηση gnp_random_graph του networkx.
def RG_conncted_percent(p, n):
    connected = 0
    for i in range(100): # create 100 graphs
        tmp_RG = nx.gnp_random_graph(n, p)
        connected = connected + nx.is_connected(tmp_RG)
    return connected / 100.0
        
    
#plot example:
def plot_conncet_RG(n, indexer):
    plt.figure(indexer)
    y = [RG_conncted_percent(p/10, n) for p in range(1,10,1)]
    print(y)
    plt.bar([i/10 for i in list(range(1,10,1))],y, width=0.05, color='b')
    plt.title("RG n = %d" %n)
    plt.ylabel("Number of connected graphs")
    plt.xlabel("Probability")
    plt.show()
    
plot_conncet_RG(100, 1)
plot_conncet_RG(200, 2)
```

Σχολιάστε τα αποτελέσματα συνεκτικότητας για το RG Gilbert:

Παρατηρούμε και εδώ 100% συνδεσιμότητα ανεξαρτήτως πλήθους ακμών και κορυφών. Αυτό οφείλεται στην ιδιότητα των τυχαίων γραφημάτων που υποστηρίζει την δημιουργία giant component πάνω στο οποίο είναι συνδεδεμένοι όλοι οι κόμβοι αν ο μέσος βαθμός των κορυφών του γραφήματος είναι μεγαλύτερος της μονάδας. Θεωρούμε πως εδώ κάτι τέτοιο ικανοποιείται ακόμη και για μικρή πιθανότητα p καθώς το πλήθος των ακμών είναι σχετικά μεγάλο σε σχέση με τις κορυφές. 

```python
#έλεγχος συνεκτικότητας RGG
def RGG_conncted_percent(R, n):
    connected = 0
    for i in range(100): # create 100 graphs
        tmp_RGG = nx.random_geometric_graph(n, R, dim = 2)
        connected = connected + nx.is_connected(tmp_RGG)
    return connected / 100.0
        
    
#plot example:
def plot_conncet_RGG(n, indexer):
    plt.figure(indexer)
    y = [RGG_conncted_percent(R/1000, n) for R in range(25,275,25)]
    print(y)
    plt.bar([i/1000 for i in list(range(25,275,25))],y, width=0.017, color='b')
    plt.title("RGG n = %d" %n)
    plt.ylabel("Number of connected graphs")
    plt.xlabel("Radius")
    plt.show()
    
plot_conncet_RGG(100, 1)
plot_conncet_RGG(200, 2)
```

Σχολιάστε τα αποτελέσματα συνεκτικότητας για το RGG:

Για το γεωμετρικό τυχαίο γράφημα σημαντικό ρόλο παίζει η ακτίνα που ορίζουμε στην αρχή, καθώς μόνο οι κόμβοι που βρίσκονται εντός της ακτίνας θα συνδεθούν μεταξύ τους. Για τον λόγο αυτό αναμένουμε ότι το ποσοστό των συνδεδεμένων γραφημάτων θα μεγαλώνει σημαντικά καθώς αυξάνεται η ακτίνα, και αντίστοιχα θα τείνει στο μηδέν καθώς η ακτίνα μειώνεται γιατί για πολύ μικρή ακτίνα η πιθανότητα να βρεθεί κόμβος εντός αυτής μειώνεται σημαντικά και έτσι συναντάμε απομονωμένους κόμβους που καθιστούν το γράφημα μη συνδεδεμένο. Τα παραπάνω επιβεβαιώνονται γραφικά καθώς όσο αυξάνουμε την ακτίνα το ποσοστό των συνδεδεμένων γραφημάτων αυξάνεται σημαντικά και σε κάποιο σημείο παρατηρούμε ραγδαία άνοδο υποδηλώνοντας την ύπαρξη κατωφλιού.

```python
#έλεγχος συνεκτικότητας SF
def SF_conncted_percent(d, n):
    connected = 0
    for i in range(100): # create 100 graphs
        tmp_SF = nx.barabasi_albert_graph(n, d)
        connected = connected + nx.is_connected(tmp_SF)
    return connected / 100.0
        
    
#plot example:
def plot_conncet_SF(n, indexer):
    plt.figure(indexer)
    y = [SF_conncted_percent(d, n) for d in range(2,12,2)]
    print(y)
    plt.bar([i for i in list(range(2,12,2))],y, width=0.5, color='b')
    plt.title("SF n = %d" %n)
    plt.ylabel("Number of connected graphs")
    plt.xlabel("Degree")
    plt.show()
    
plot_conncet_SF(100, 1)
plot_conncet_SF(200, 2)
```

Σχολιάστε τα αποτελέσματα συνεκτικότητας για το SF:

Για τα scale free γραφήματα αναμέναμε να είναι συνδεδεμένα ανεξαρτήτως βαθμού καθώς σε αυτά υπάρχουν κόμβοι μεγάλου βαθμού πάνω στους οποίους είναι συνδεδεμένοι πολλοί κόμβοι μικρού βαθμού.

```python
# για την κατασκευή της SW, χρησιμοποιήστε την connected_watts_strogatz_graph() 
def SW_conncted_percent(d, n, p):
    connected = 0
    for i in range(100): # create 100 graphs
        tmp_SW = nx.connected_watts_strogatz_graph(n, d, p)
        connected = connected + nx.is_connected(tmp_SW)
    return connected / 100.0
        
    
#plot example:
def plot_conncet_SW(n, indexer):
    plt.figure(indexer)
    y = [SW_conncted_percent(d, n, p/10) for d in range(2,12,2) for p in range(1,8,1)]
    print(y)
    plt.bar([(i+2*j/10) for i in list(range(2,12,2)) for j in list(range(1,8,1))],y, width=0.05, color='b')
    plt.title("SW n = %d" %n)
    plt.ylabel("Number of connected graphs")
    plt.xlabel("Degree and probability")
    plt.show()
    
plot_conncet_SW(100, 1)
plot_conncet_SW(200, 2)
```

Για να το κάνουμε αυτό θα υποθέσουμε ότι $n^2 \cdot p \rightarrow \infty$ υπό αυτή τη συνθήκη και με βάση τον νόμο των μεγάλων αριθμών ξέρουμε ότι για το G(n, p) το αναμενόμενο πλήθος ακμών του θα είναι ${n\choose2}p$. Για να έχουμε ισοδυναμία των δύο μοντέλων απαιτούμε αυτό το πλήθος στο ένα μοντέλο να ισούται με το Μ του άλλου και αναλόγως τα δεδομένα επιλύουμε την σχέση. Από το νόμο των μεγάλων αριθμών παρατηρούμε ότι το παραπάνω θα ισχύει και τα δίκτυα θα συμπεριφέρονται παρόμοια υπό αυτή τη συνθήκη.

Λύνοντας αριθμητικά το Μ = ${n\choose2}p$ προκύπτουν τα παραπάνω νούμερα.

```python
def ego_central_caclulator(G):
    result = []
    for i in range(n):
        ego = nx.ego_graph(G,i)
        adj_mat = (nx.adjacency_matrix(ego)).todense()
        #print(adj_mat)
        hadamard_product = np.multiply(np.linalg.matrix_power(adj_mat, 2), (1 - adj_mat))
        #print(hadamard_product)
        up_triangular = np.triu(hadamard_product, 1)
        #print(up_triangular)
        non_zero = list(up_triangular[np.nonzero(up_triangular)])
        #print(non_zero)
        half_sum_of_recipr = sum(1/i for i in non_zero) / 2
        result.append(half_sum_of_recipr)
    return result

REG =  nx.watts_strogatz_graph(n, d, 0)
RGER = nx.gnm_random_graph(n, M)
RGG = nx.random_geometric_graph(n, R, dim = 2)
SF = nx.barabasi_albert_graph(n, d)
SW = nx.watts_strogatz_graph(n, d, p) # remake graphs without weights

#one_matrix = np.ones((5,5))
ego_REG = ego_central_caclulator(REG)
ego_RGER = ego_central_caclulator(RGER)
ego_RGG = ego_central_caclulator(RGG)
ego_SF = ego_central_caclulator(SF)
ego_SW = ego_central_caclulator(SW)

print('REG egocentrality: ', ego_REG, '\n')
print('RGER egocentrality: ', ego_RGER, '\n')
print('RGG egocentrality: ', ego_RGG, '\n')
print('SF egocentrality: ', ego_SF, '\n')
print('SW egocentrality: ', ego_SW, '\n')
```
