### Άσκηση 1: Εξοικείωση με Γενετικούς Αλγόριθμους

#### Πρόβλημα ONEMAX

<p style=text-align:justify>Βρείτε την δυαδική ακολουθία $(x_1, x_2, ..., x_n)$ που μεγιστοποιεί το άθροισμα $x_1 + x_2 + ... + x_n, n=20$. Αναλύστε την επίδραση του μεγέθους του πληθυσμού, της πιθανότητας διασταύρωσης (crossover) και της πιθανότητας μετάλλαξης (mutation) στην ποιότητα της λύσης (πόσο κοντά είναι στην προφανή βέλτιστη λύση) που προκύπτει από το γενετικό αλγόριθμο. Συγκεκριμένα, μεταβάλλετε τον πληθυσμό από $10$ ως $100$ χρωμοσώματα με βήμα $10$,την πιθανότητα διασταύρωσης από $0.3$ ως $0.9$ με βήμα $0.1$ και την πιθανότητα μετάλλαξης από $0.01$ ως $0.2$ με βήμα $0.01$. Εξετάστε την ποιότητα της λύσης του γενετικού αλγορίθμου καθώς αλλάζετε τις παραμέτρους. Ελιτισμό μπορείτε να χρησιμοποιήσετε αν το επιθυμείτε.</p>

```python
def create_population(num): # num is population number
    res = []
    for i in range(num):
        chromo = [np.random.randint(0,2) for i in range(20)] # create chromosome of len 20
        res.append(chromo) #append chromo to population
    # Create the initial population
    return res

def find_fitness(chromo): #fitness of each chromo
    return (sum(chromo)) # fitness is the sum of chromo's bits
    # Calculate fitness of chromosomes


def proportional_selection(chromo_list, num, elitism):
    all_fitnesses = [find_fitness(i) for i in chromo_list] #fitness for all chromos
    total_fitness = sum(all_fitnesses)
    new_population = []
    
    sorted_fit = [(find_fitness(i), i) for i in chromo_list]
    sorted_fit.sort(key = lambda x: x[0], reverse = True) #sort according to fitness in order to keep the best
    for i in range(elitism):
        new_population.append(sorted_fit[i][1]) #append best chromos
    
    for pop in range(num - elitism): #select #num of chromos with roulette
        x = random.uniform(0,1) # random num in [0,1]
        i = 0 # start from the beginning of list
        cur_sum = all_fitnesses[i] / total_fitness
        while(i < num and x > cur_sum):
            i += 1
            cur_sum += all_fitnesses[i] / total_fitness #update sum
        new_population.append(chromo_list[i])
    return new_population
    # Select chromosomes to pass to next population proportionally to their fitness


def crossover(chromo_list, prob): #prob = cross probability
    new_population = []
    while(len(new_population) < len(chromo_list)):
        for i in range(0, 2, len(chromo_list)): #run for every chromo pair
            x = random.uniform(0,1) # prob to cross
            if(x <= prob): #if within given prob then cross
                x = random.randint(0, len(chromo_list[0])) # cross point selected at random
                new_chromo = chromo_list[i][:x] #first x bits
                new_chromo.extend(chromo_list[i + 1][x:]) #last n-x bits
                new_population.append(new_chromo)
        # Apply crossover to population
    return new_population

def mutation(chromo_list, prob):
    for i in range(len(chromo_list)):
        for j in chromo_list[i]: # for each gene in chromo
            x = random.uniform(0,1)
            if (x < prob): # mutation happened
                j = 1 - j #flip bit
    return chromo_list
    # Apply mutation to population
    

def genetic_algorithm(pop_num, cross_prob, mut_prob, elitism):
    # Main genetic algorithm
    init_pop = create_population(pop_num)
    top_scores = []
    for jj in range(30):
        selected_pop = proportional_selection(init_pop, len(init_pop), elitism)
        crossed_pop = crossover(selected_pop, cross_prob)
        mutated_pop = mutation(crossed_pop, mut_prob)
        init_pop = mutated_pop
        top_fitness = max([find_fitness(i) for i in mutated_pop])
        top_scores.append(top_fitness)
        if(len(top_scores) >= 5): #check if last 5 iterations where the same best score
            if(top_scores[-1] == top_scores[-2] == top_scores[-3] == top_scores[-4] == top_scores[-5]):
                break #break loop if last 5 iterations had same best score
        
    return mutated_pop
```

```python
pop_list = [i for i in range(10, 210, 10)]
cross_list = list(np.linspace(0.3, 0.9, 7))
mut_list = list(np.linspace(0.01,0.2,20))
elitism = [i for i in range(0, 3)]

best_chromos = []
for i in pop_list:
    for j in cross_list:
        for k in mut_list:
            for l in elitism:
                xx = genetic_algorithm(i, j, k, l)
                best_chromos.append([i, j, k, l, xx[np.argmax([find_fitness(ii) for ii in xx])]])
```

### Άσκηση 2: Εντοπισμός Κοινοτήτων σε Γράφους Κοινωνικών Δικτύων με Χρήση Γενετικών Αλγορίθμων

<p style=text-align:justify>Στην παρούσα άσκηση θα αναπτύξετε κώδικα για εντοπισμό κοινοτήτων σε γράφους δικτύων με χρήση γενετικών αλγορίθμων. Ακολούθως θα εφαρμόσετε το γενετικό αλγόριθμο για τον εντοπισμό κοινοτήτων σε πραγματικά δίκτυα και θα συγκρίνετε το αποτέλεσμά του με εκείνο των αλγορίθμων της δεύτερης εργαστηριακής άσκησης χρησιμοποιώντας τη μετρική modularity.</p>

##### Αναπαράσταση υποψήφιων λύσεων (χρωμοσωμάτων)
<p style=text-align:justify>Κάθε χρωμόσωμα $b_{i,t}$ του πληθυσμού τη χρονική στιγμή $t$ είναι ένα διάνυσμα με μήκος ίσο με το πλήθος των κόμβων του δικτύου, $n$. Αν $b_{i,t}(j)=d$ τότε οι κόμβοι $j,d$ ανήκουν στην ίδια κοινότητα. Επιπλέον για να ισχύει $b_{i,t}(j)=d$ , οι κόμβοι $j,d$ θα πρέπει να συνδέονται στην αρχική τοπολογία. Τότε, βρίσκοντας τις συνδεδεμένες συνιστώσες του χρωμοσώματος $b_{i,t}$ , προκύπτουν οι κοινότητες του δικτύου.</p>

##### Αρχικοποίηση
<p style=text-align:justify>Η αρχική γενιά του πληθυσμού θα πρέπει να αποτελείται από χρωμοσώματα τα οποία θα είναι τυχαία επιλεγμένα και διορθωμένα ώστε ο αριθμός που υπάρχει σε κάθε θέση να είναι γείτονας του αντίστοιχου κόμβου. (Αν δεν ισχύει θα πρέπει να αντικατασταθεί από ένα γείτονα του αντίστοιχου κόμβου.) Θεωρείστε πληθυσμό $300$ χρωμοσωμάτων.</p>

##### Συνάρτηση Fitness
<p style=text-align:justify>Η συνάρτηση fitness υπολογίζεται όπως στις διαφάνειες 28 και 29 της αντίστοιχης διάλεξης του μαθήματος. Προσοχή για κάθε συνιστώσα (component) του γράφου θα πρέπει να βρίσκετε τον αντίστοιχο υπογράφο.</p>

##### Επιλογή (Selection)
<p style=text-align:justify>Η επιλογή χρωμοσωμάτων από μία γενιά του πληθυσμού για την κατασκευή της επόμενης γενιάς γίνεται με τη μέθοδο της ρουλέτας (διαφάνειες μαθήματος και αλγόριθμος). Τροποποιήστε τον παραπάνω αλγόριθμο ώστε να εφαρμόσετε και ελιτισμό, δηλαδή τα $x$ πρώτα χρωμοσώματα της νέας γενιάς θα είναι εκείνα τα χρωμοσώματα της προηγούμενης γενιάς που έχουν επιτύχει τη μέγιστη τιμή της συνάρτησης fitness.</p>

##### Διασταύρωση (Crossover)
<p style=text-align:justify>Η διασταύρωση πραγματοποιείται όπως περιγράφεται στον αλγόριθμο, μεταξύ διαδοχικών χρωμοσωμάτων της εκάστοτε γενιάς. Η πιθανότητα διασταύρωσης ανά ζεύγος είναι ίση με $p_C$.</p>

##### Μετάλλαξη (Mutation)
<p style=text-align:justify>Η μετάλλαξη πραγματοποιείται σε κάθε θέση του χρωμοσώματος με πιθανότητα $p_M$, όπως στον αλγόριθμο. Θα πρέπει ωστόσο ο αλγόριθμος να τροποποιηθεί ώστε να δίνει ένα τυχαίο γείτονα του κόμβου που αντιστοιχεί σε κάθε θέση ως αποτέλεσμα της μετάλλαξης.</p>

##### Κριτήριο Τερματισμού (Stopping Criterion)

```python
#### ATTENTION: don't change because the variables are used in the given code
### the use of the converted graph is optional

# Import real topologies 

football_gml = nx.read_gml("./football.gml")
lesmis_gml = nx.read_gml("./lesmis.gml")
dolphins_gml = nx.read_gml("./dolphins.gml")

def conv2int(G,start_value):
    nG = nx.convert_node_labels_to_integers(G, first_label=start_value)
    G_mapping = dict(enumerate(G.nodes,0))
    return nG,G_mapping

football, football_map = conv2int(football_gml,0)
lesmis, lesmis_map = conv2int(lesmis_gml,0)
dolphins, dolphins_map = conv2int(dolphins_gml,0)


# i=[0,3,5]
# row = nx.adjacency_matrix(football).todense()[11].tolist() #find rows representing component nodes
# print(row, len(row[0]), row[0][3], [row[0][rr] for rr in range(len(row[0]))])
# final_row = [row[0][rr] for rr in range(len(row[0])) if rr in i]
# #ress.append((nx.adjacency_matrix(football).todense()[11]).tolist())
# print(final_row)
# print(list(nx.Graph.neighbors(football, 114)))

# [len(c) for c in sorted(nx.connected_components(football), key=len, reverse=True)]


```

```python
def create_population(G): # chromo size = n, G is graph
    # Create the initial population
    chromo_size = len(list(G.nodes))
    res = np.zeros((300, chromo_size))
    neighbors1 = []
    for i in range(len(G.nodes())):
       neighbors1.append(list(G.neighbors(i))) 
    
    for i in range(300): # create 300 chromos
        for j in range(chromo_size):
            res[i][j] = np.random.choice(neighbors1[j], 1)[0]
    return res.astype(int)


def find_fitness(chromo, G):
    # Calculate fitness of chromosomes
    adja_mat = np.squeeze(np.asarray(nx.to_numpy_array(G)))#nx.adjacency_matrix(G).todense()
    chromo_size = len(list(G.nodes))
    tem_G = nx.Graph() #create chromo graph in order to find components
    tem_G.add_nodes_from([i for i in range(chromo_size)]) #add all nodes
    tem_G.add_edges_from([(i, chromo[i]) for i in range(len(chromo))]) #add edges using chromo
    all_components = [list(i) for i in list(nx.connected_components(tem_G))] # list of lists containing nodes in ever connected component
    #calculate total fitness by iterating over components
    CS = 0
    #print(adja_mat)
    for i in all_components:
        S = (adja_mat)[i][:,i]
        u_s = sum(sum(S))
        M_s = sum(np.mean(S, axis = 1)) / (S.shape[0])
        CS += M_s * u_s #increase CS accordingly the calculated values
    #print(S)
    return CS #community score of given chromo
        
        

def proportional_selection(chromo_list, num, elitism, G):
    all_fitnesses = [find_fitness(i, G) for i in chromo_list.tolist()] #fitness for all chromos
    total_fitness = sum(all_fitnesses)
    new_population = np.zeros((300, len(G.nodes())))
    
    x_elits = np.argsort(all_fitnesses)[::-1][0:elitism]
    new_population[0:elitism] = chromo_list[x_elits]  
    
    for pop in range(elitism, num): #select #num of chromos with roulette
        x = random.uniform(0,1) # random num in [0,1]
        i = 0 # start from the beginning of list
        cur_sum = all_fitnesses[i] / total_fitness
        while(i < num and x > cur_sum):
            i += 1
            cur_sum += all_fitnesses[i] / total_fitness #update sum
        new_population[pop,:] = np.reshape(chromo_list[i,:], (1,len(G.nodes())))
    return new_population
    # Select chromosomes to pass to next population proportionally to their fitness
    

def crossover(chromo_list, prob):
    # Apply crossover to population
    new_population = np.zeros((300, chromo_list.shape[1]))
    mm = 0
    while(mm < len(chromo_list)):
        for i in range(0, len(chromo_list)-1): #run for every chromo pair
            x = random.uniform(0,1) # prob to cross
            if(x <= prob): #if within given prob then cross
                mask = [np.random.randint(0,2) for ij in range(len(chromo_list[0]))] #mask of size of chromo
                new_chromo = [chromo_list[i][kk] if mask[kk] == 0 else chromo_list[i + 1][kk] for kk in range(len(mask))] #create new chromo combining parents
                new_population[mm] = (new_chromo)
                mm += 1
                if(mm == 300):
                    break
        # Apply crossover to population
    return new_population
    

def mutation(chromo_list, prob, G):
    chromo_size = len(list(G.nodes))
    
    neighbors1 = []
    for i in range(len(G.nodes())):
       neighbors1.append(list(G.neighbors(i))) 
    
    for i in range(len(chromo_list)):
        for j in range(len(chromo_list[i])): # for each gene in chromo
            x = random.uniform(0,1)
            if (x < prob): # mutation happened#use random neighbor
                chromo_list[i][j] = np.random.choice(neighbors1[j], 1)[0]
    return chromo_list.astype(int)
    # Apply mutation to population

import time
def genetic_algorithm(G, cross_prob, mut_prob, elitism):
    # Main genetic algorithm
    init_pop = create_population(G)
    top_scores = []
    for jj in range(30):
        selected_pop = proportional_selection(init_pop, len(init_pop), elitism, G)
        crossed_pop = crossover(selected_pop, cross_prob)
        mutated_pop = mutation(crossed_pop, mut_prob, G)
        init_pop = mutated_pop
        ##### SECOND STOP CONDITION #####
        top_fitness = max([find_fitness(i, G) for i in mutated_pop.tolist()])
        top_scores.append(top_fitness)
        if(len(top_scores) >= 5): #check if last 5 iterations where the same best score
           if(top_scores[-1] == top_scores[-2] == top_scores[-3] == top_scores[-4] == top_scores[-5]):
               break #break loop if last 5 iterations had same best score
        
    return mutated_pop
```

```python
# Run the genetic algorithm for different values of the requested parameters
cross_list = list(np.linspace(0.7, 0.9, 3))
mut_list = [0.1, 0.2]
elitism = [i for i in range(1, 4)]


def checker_GA(G): #return best chromo of every combination
    res = []
    indexers = []
    for i in cross_list:
        for j in mut_list:
            for k in elitism:
                pop = genetic_algorithm(G, i, j, k) #population for given values
                best_chromo = pop[np.argmax([find_fitness(vv, G) for vv in pop.tolist()])]
                res.append(best_chromo) #keep only best chromo of population
                indexers.append([find_fitness(best_chromo, G), i, j, k])
    
    indexers.sort(key = lambda x: x[0], reverse = True)
    print('Best parameters for Fitness = {} are: Pc: {} Pm: {} Elitism: {}'.format(indexers[0][0], indexers[0][1], indexers[0][2],indexers[0][3]))
    return res # return all best chromos of given graph for every value combination

print('Football')
football_population = checker_GA(football) # create population of best chromos for all possible combinations
print('Lesmis')
lesmis_population = checker_GA(lesmis)
print('Dolphins')
dolphin_population = checker_GA(dolphins)
```

```python
#best chromo of given parameters
#find best chromo for all parameters and feed it to compute_modularity
#one with best performance is the best result of GA

def communities_analyzer(best_chromo): #returns communities for give chromo in proper way
    tem_G = nx.Graph() #create chromo graph in order to find components
    tem_G.add_nodes_from([i for i in range(len(best_chromo))]) #add all nodes
    tem_G.add_edges_from([(i, best_chromo[i]) for i in range(len(best_chromo))]) #add edges using chromo
    all_components = [list(i) for i in list(nx.connected_components(tem_G))] # list of lists containing nodes in ever connected component
    return all_components 
```

```python
# modularity for GA community detection algorithm 
def compute_modularity(G,G_results): #G_results is community partion
    res = []
    for i in G_results: # for every possible community find the best
        #num_of_communities = len(i)
        com = communities_analyzer(i)
        num_of_communities = len(com)
        perform = nx.algorithms.community.quality.performance(G, com)
        res.append((i, perform, num_of_communities)) #list with all communities and their features 
    res.sort(key = lambda x: x[1], reverse = True) #sort according to performance
    return res[0] #return best community partition
    


football_GA, football_GA_mod, football_GA_num = compute_modularity(football, football_population)
lesmis_GA, lesmis_GA_mod, lesmis_GA_num = compute_modularity(lesmis, lesmis_population)  
dolphins_GA, dolphins_GA_mod, dolphins_GA_num = compute_modularity(dolphins, dolphin_population)
```

### Find the best partition with Girvan-Newman, Spectral Clustering and Modularity Maximization
<p style=text-align:justify>Aποτέλεσματα των αλγορίθμων της δεύτερης εργαστηριακής άσκησης (με προτεινόμενη υλοποίηση).</p>

```python
import itertools

# Girvan_newman:partitions of at most k communities, select the one with the highest value of modularity.
def comd_GN(G,k):
    gn_G = nx.algorithms.community.centrality.girvan_newman(G)
    limited = itertools.takewhile(lambda c: len(c) <= k, gn_G)
    per_G = []
    com_G = []
    for communities in limited:
        per = nx.algorithms.community.quality.performance(G,communities)
        per_G.append(per) #list of performance of each partition
        com_G.append(communities) # list of partitions
    max_mod_G = max(per_G) # maximum modularity value
    index_G = per_G.index(max(per_G)) # index of maximum modularity
    best_G = com_G[index_G] # the partition of maximum modularity  
    num_best_G = len(best_G)
    return best_G, max_mod_G, num_best_G

GN_best_football, GN_mod_football, GN_num_football = comd_GN(football,50)
GN_best_lesmis, GN_mod_lesmis, GN_num_lesmis = comd_GN(lesmis,50)
GN_best_dolphins, GN_mod_dolphins, GN_num_dolphins = comd_GN(dolphins,50)
```

```python
from sklearn.cluster import SpectralClustering

# Spectral Clustering
def comd_SP(G,k):
    mG = np.array(nx.to_numpy_matrix(G))
    clustering = SpectralClustering(n_clusters=k,random_state=0, affinity='precomputed').fit(mG)
    clabels = clustering.labels_
    numc = clustering.n_clusters
    b = dict(enumerate(clabels,0)) # create a mapping of nodes indices to the community it belongs
    d = [] # a list of all communities (each list is a community)
    for i in range(0,numc):
        d.append([k for k,v in b.items() if float(v) == i])
    per = nx.algorithms.community.quality.performance(G,d)
    return d, per, numc

def find_best_SP_partition(G,k):
    com = []
    com_mod = []
    com_num = []
    for i in range(2,k):
        tcom,tcom_mod,tcom_num = comd_SP(G,i)
        com.append(tcom)
        com_mod.append(tcom_mod)
        com_num.append(tcom_num)
    index = com_mod.index(max(com_mod))
    G_SP = com[index]
    G_SP_mod = com_mod[index]
    G_SP_num = com_num[index]
    return G_SP, G_SP_mod, G_SP_num

football_SP, football_SP_mod, football_SP_num = find_best_SP_partition(football,50) 
lesmis_SP, lesmis_SP_mod, lesmis_SP_num = find_best_SP_partition(lesmis,50)  
dolphins_SP, dolphins_SP_mod, dolphins_SP_num = find_best_SP_partition(dolphins,50)
```

```python
# Modularity Maximization
def comd_MM(G):
    c = list(nx.algorithms.community.modularity_max.greedy_modularity_communities(G))
    mm_G = []
    for i,j in enumerate(c):
        a = list(j)
        mm_G.append(a)
    per = nx.algorithms.community.quality.performance(G,mm_G)
    num_G = len(mm_G)
    return mm_G, per,num_G

football_MM, football_MM_mod, football_MM_num = comd_MM(football)
lesmis_MM, lesmis_MM_mod, lesmis_MM_num = comd_MM(lesmis)  
dolphins_MM, dolphins_MM_mod, dolphins_MM_num = comd_MM(dolphins)
```
